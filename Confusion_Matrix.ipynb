```python
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

np.random.seed(0)
#intialize two dummy lists one for before prediction ,other for after prediction
target = np.random.randint(low=0,high=2,size=10)
predicted = np.random.randint(low=0,high=2,size=10)

# print both the list
print("target=",target,"\n")
print("predicted=",predicted,"\n")

#using scikit learn library
print("Sklearn Confusion Matrix:",confusion_matrix(target,predicted),sep="\n")
```

```python
def confusion_matrix(target,predicted):
    TP=0
    TN=0
    FP=0
    FN=0
    for actual,pred in zip(target,predicted):
        if actual==1 and pred==1:
            TP += 1
        elif actual==0 and pred==0:
            TN += 1
        elif actual==1 and pred==0:
            FN += 1
        elif actual==0 and pred==1:
            FP += 1
    #compute scores after prediction
    sensitivity=(TP/(TP+FN))
    specificity=(TN/(TN+FP))
    precision=(TP/(TP+FP))
    F1_score=(2*(sensitivity*precision))/(precision+sensitivity)
    accuraccy=(TP+TN)/((TP+FN)+(TN+FP))
    
    #print("\nconfusion matrix witout scikit learn library")
    print("sensitivity/Recall=",sensitivity)
    print("specificity=",specificity)
    print("precision=",precision)
    print("F1-score=",F1_score)
    print("accurracy=",accuraccy)
    return np.array([[TN,FP],
                     [FN,TP]])

def main():
    #print("Custom Confusion Matrix:",confusion_matrix(target,predicted),sep="\n")
    cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(target,predicted), display_labels = [0, 1])
    cm_display.plot()
    plt.show()
    matrix = classification_report(target,predicted,labels=[1,0])
    print('Classification report : \n',matrix)

    
main()
    
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
actual    = np.array(['Dog','Dog','Dog','Not Dog','Dog','Not Dog','Dog','Dog','Not Dog','Not Dog'])
predicted = np.array( ['Dog','Not Dog','Dog','Not Dog','Dog','Dog','Dog','Dog','Not Dog','Not Dog'])
```

```python
print("length of actual=",len(actual))
print("length of predicted=",len(predicted))
```

```python
cm = confusion_matrix(actual,predicted)
print(cm)
```

```python
sns.heatmap(cm, 
            annot=True,
            fmt='g', 
            xticklabels=['Dog','Not Dog'],
            yticklabels=['Dog','Not Dog'])
plt.ylabel('Actual', fontsize=12)
plt.title('Confusion Matrix', fontsize=15, pad=20)
plt.gca().xaxis.set_label_position('top') 
plt.xlabel('Prediction', fontsize=12)
plt.gca().xaxis.tick_top()
plt.gca().figure.subplots_adjust(bottom=0.2)
plt.gca().figure.text(0.8, 0.08, 'Prediction', ha='center', fontsize=12)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

actual = [1,0,0,1,0,0,1,0,0,1]
# predicted values
predicted = [1,0,0,1,0,0,0,1,0,0]

# confusion matrix
matrix = confusion_matrix(actual,predicted, labels=[1,0])
print('Confusion matrix : \n',matrix)

# outcome values order in sklearn
tp, fn, fp, tn = confusion_matrix(actual,predicted,labels=[1,0]).reshape(-1)
print('Outcome values : \n', tp, fn, fp, tn)

# classification report for precision, recall f1-score and accuracy
matrix = classification_report(actual,predicted,labels=[1,0])
print('Classification report : \n',matrix)
```

```python
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

actual = numpy.random.binomial(1,.9,size = 1000)
predicted = numpy.random.binomial(1,.9,size = 1000)
```

```python
print(actual)
print(predicted)
```

```python
confusion_matrix = metrics.confusion_matrix(actual, predicted)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])

cm_display.plot()
plt.show()
```

```python
Accuracy = metrics.accuracy_score(actual, predicted)
print(Accuracy)
```

```python
Precision = metrics.precision_score(actual, predicted)
print(Precision)
```

```python
Sensitivity_recall = metrics.recall_score(actual, predicted)
print(Sensitivity_recall)
```

```python
Specificity = metrics.recall_score(actual, predicted, pos_label=0)
print(Specificity)
```

```python
F1_score = metrics.f1_score(actual, predicted)
print(F1_score)
```

```python
#metrics
print({"Accuracy":Accuracy,"Precision":Precision,"Sensitivity_recall":Sensitivity_recall,"Specificity":Specificity,"F1_score":F1_score})
```
